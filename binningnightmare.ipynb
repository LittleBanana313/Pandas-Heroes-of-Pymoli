## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# create bins and associated labels
bins = (0, 9, 14, 19, 24, 29, 34, 39, 200)
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
# categorize players into bins and remove duplicate screen names
agedf = PD_df
agedf['Total Count'] = pd.cut(agedf["Age"], bins, labels=labels)
dfage = PD_df[['Total Count','SN']].drop_duplicates(subset = 'SN')
agedf
# group by bins find percentages
agesum = dfage.groupby('Total Count').count()
Acount = agesum['SN']
Pcount = len(pd.unique(PD_df["SN"]))
Percage = round(Acount / Pcount * 100, 2).astype(str) + '%'
# transpose and print dataframe
df1 = pd.DataFrame([Acount, Percage])
df2 = df1.T
df2.columns = ['Total Number', 'Percentage of Players']
df3 = df2.rename_axis('Age Range', axis='columns')
# df3.drop('Total Count', axis=0)
df3
```
