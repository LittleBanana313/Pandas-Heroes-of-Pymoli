```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
PD_df = pd.DataFrame(purchase_data)
PD_df
```

## Player Count

```python
# Find total players and make a DataFrame display
Pcount = len(pd.unique(PD_df["SN"]))
# len(Pcount)
Numdf = pd.DataFrame({"Total Players": [Pcount]})
Numdf.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
# Assign variables for use in the dataframe
Uitem = len(pd.unique(PD_df["Item Name"]))
avg_p = (PD_df["Price"].mean())
numpur = len(PD_df["Purchase ID"].value_counts())
totalp = float(PD_df["Price"].sum())

# Load values into Dataframe and display
panalysis = pd.DataFrame({"Number of Unique Items": [Uitem], "Average Price":[avg_p], "Number of Purchases": [numpur], "Total Revenue": [totalp]})
panalysis.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# group using gender as key
gender = PD_df.groupby("Gender")
# Count names by gender
countgender = gender.nunique()["SN"]
#formula for percentages
percentG = round((countgender/Pcount)*100).astype(str) + '%'
# create gender dataframe 
gender_df = pd.DataFrame({"Total Count": countgender,"Percentage of Players": percentG})
gender_df.index.name = None
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# formulate variables by gender
Pcount = round(gender["Price"].count(), 0)
Aprice = round(gender["Price"].mean(), 2)
Tprice = round(gender["Price"].sum(), 2) 
AvgbyG = round(Tprice/countgender, 2)

# input variables into DataFrame, transpose and print
df = pd.DataFrame([Pcount, Aprice, Tprice, AvgbyG])
df1 = df.T
df1.columns = ["Purchase Count", "Average Purchase Price", 
               "Total Purchase Value", 'Avg Total Purchase per Person']
df1
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# create bins and associated labels
bins = (0, 9, 14, 19, 24, 29, 34, 39, 200)
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
# categorize players into bins and remove duplicate screen names
agedf = PD_df
agedf['Total Count'] = pd.cut(agedf["Age"], bins, labels=labels)
dfage = PD_df[['Total Count','SN']].drop_duplicates(subset = 'SN')
agedf
# group by bins find percentages
agesum = dfage.groupby('Total Count').count()
Acount = agesum['SN']
Pcount = len(pd.unique(PD_df["SN"]))
Percage = round(Acount / Pcount * 100, 2).astype(str) + '%'
# transpose and print dataframe
df1 = pd.DataFrame([Acount, Percage])
df2 = df1.T
df2.columns = ['Total Number', 'Percentage of Players']
df3 = df2.rename_axis('Age Range', axis='columns')
df3 = df3.rename_axis(' ', axis='rows')
df3
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# create bins and associated labels
bins = (0, 9, 14, 19, 24, 29, 34, 39, 200)
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
# create dataFrame for grouping
agedf1 = PD_df
agedf1['Total Count'] = pd.cut(agedf1["Age"], bins, labels=labels)
Lrows = agedf1[['Total Count','SN', 'Price']]
rowsG = Lrows.groupby('Total Count')

# calculate variables
Pcount2 = (rowsG['Price'].count)
tpv = round(rowsG['Price'].sum(), 2)
avgP = round(rowsG['Price'].mean(), 2)
Spendper = round(tpv/Acount, 2)

# clean up column headers and print dataframe
df4 = pd.DataFrame({'Purchase Count': Pcount2, 'Average Purchase Price': avgP,
                    'Total Purchase Value': tpv, 'Avg Total Purchase per Person': Spendper})
df4 = df4.rename_axis('Age Range', axis='columns')
df4 = df4.rename_axis(' ', axis='rows')
df4
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# set up grouping
spend_df = PD_df.groupby('SN')
# use grouped dataframe to formulate variables
avgprice = round(spend_df['Price'].mean(), 2)
count = spend_df['SN'].count()
aspend = round(spend_df['Price'].mean(),2)
tp2 = round(spend_df['Price'].sum(),2)
# use one dataframe to store data, second to transpose and name columns
topdf = pd.DataFrame([count, aspend, tp2])
topdf2 = topdf.T
topdf2.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']
# use second dataframe to correct funky columns and sort values
topdf2 = topdf2.rename_axis('Screen Name', axis='columns')
topdf2 = topdf2.rename_axis(' ', axis='rows')
topdf2.sort_values('Total Purchase Value', ascending=False).head(5)
```
